# Data Import & Exploration

```python
import pandas as pd
```

### Inspect and clean financial columns

```python
path = r"/content/drive/MyDrive/Revision tech/financial data sp500 companies.csv"

df = pd.read_csv(path, sep=',', encoding='utf-8')   # ← essaie d’abord ce séparateur
df.shape
df.info()
df.head(3)
df.isnull().sum().sort_values(ascending=False).head(10)
df.describe()
```

```python
df=df.drop(columns=['Unnamed: 0'],errors='ignore')
df.columns=(df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^a-z0-9_]','',regex=True))

df.head(5)
```

```python
df['date']=pd.to_datetime(df['date'],errors='coerce')
# Liste des colonnes numériques
cols_num=df.select_dtypes(include='number').columns.tolist()

# Si certaines colonnes sont encore "object", forcer la conversion
df[cols_num] = df[cols_num].apply(pd.to_numeric, errors='coerce')
```

```python
missing= df.isnull().sum()
missing_pct=(missing/len(df))*100
print(missing_pct.sort_values(ascending=False))
```

```python
df['research_development']=df['research_development'].fillna(0)
```

```python
df=df.dropna(subset=['date'])
```

```python
df['interest_expense']=df['interest_expense'].fillna(0)
```

```python
df['selling_general_administrative'] = df['selling_general_administrative'].fillna(
    df['selling_general_administrative'].median()
)
```

```python
for col in ['net_income_applicable_to_common_shares','operating_income','income_before_tax','net_income_from_continuing_ops']:
  df[col]=df[col].fillna(df[col].median())
```

```python
missing= df.isnull().sum()
missing_pct=(missing/len(df))*100
print(missing_pct.sort_values(ascending=False))
```

```python
print(df.describe().T[['mean', 'min', 'max']])
```

```python
import numpy as np

# 1️ R&D can't be negative
df['research_development'] = df['research_development'].clip(lower=0)

# 2️ Interest Expense : has to be negative (charge financière)
df.loc[df['interest_expense'] > 0, 'interest_expense'] *= -1

# 3️ Check that Total Revenue and Gross Profit are positive
df = df[(df['total_revenue'] > 0) & (df['gross_profit'] > 0)]

# 4️ Vérifie cohérence logique entre postes
df = df[df['income_before_tax'].notna()]
df = df[df['net_income'].notna()]

# 5️ Delete all impossible values (revenues < Total negative depenses)
df = df[(df['total_operating_expenses'] > 0)]

# 6️ Some ratio to Check
df['net_margin'] = df['net_income'] / df['total_revenue']
df['operating_margin'] = df['operating_income'] / df['total_revenue']
df['tax_rate'] = df['income_tax_expense'] / df['income_before_tax']

# 7️ Caping the ratios (> 1 or < -1)
df['net_margin'] = df['net_margin'].clip(-1, 1)
df['operating_margin'] = df['operating_margin'].clip(-1, 1)
df['tax_rate'] = df['tax_rate'].clip(0, 1)

# 8️ Final Check
print(" Shape finale :", df.shape)
print(df[['net_margin', 'operating_margin', 'tax_rate']].describe().round(3))
```

```python
df.to_csv("/content/drive/MyDrive/Revision tech/sp500_clean.csv", index=False)
print("Clean DB saved !")
```
